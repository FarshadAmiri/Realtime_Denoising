```python
from pydub import AudioSegment
import os; os.chdir(r"C:\Users\User_1\Desktop\speeches\1")
```

```python
# List of your input files
files = [
    "3000-1.flac", "422-1.flac", "777-1.flac", "422-2.flac",
    "3000-2.flac", "3000-3.flac", "777-2.flac", "1993-1.flac", "1993-2.flac",
    "3000-4.flac",
]
```

```python
# Load and concatenate
combined = AudioSegment.empty()
for f in files:
    audio = AudioSegment.from_file(f)
    combined += audio

# Export the result
combined.export("concatenated.wav", format="wav")
print("✅ Saved as concatenated.wav")
```
